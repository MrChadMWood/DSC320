## Riemann Sum Function
<b>A Riemann sum is an estimation of the area under a curve using rectangles. The right Riemann sum <i>R<sub>n</sub></i> of a function <i>f</i> on the interval [<i>a, b</i>] with n rectangles is calculated by</b>

<b> Write a Python function to take in a mathematical function <i>f</i>, two interval endpoints <i>a</i> and <i>b</i>, and the number of rectangles <i>n</i>, and then outputs the right Riemann sum R<sub>n</sub>.</b>

```python
import math

def calcu(function: str, a, b, n):
    def f(x):
        return eval(function)
    
    area = 0 # Accumulates area through for loop
    deltaX = (b-a)/n # Constant: x
    for rect in range(n):
        rect += 1 # Compensates range() noninclusive
        x = a+(rect*deltaX) # Determines (a + nx) as x
        area += f(x)*deltaX # Determines f(x)*x and accumulates area
        
    return area
```

```python
# Test using problem provided by Khan Academy 
area = calcu('x**3 + 2', -1, 5, 3)
area # Correct answer: 318
```
